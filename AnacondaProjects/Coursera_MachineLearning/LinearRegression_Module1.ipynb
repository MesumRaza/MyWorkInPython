```python
linear_model=lm.LinearRegression()
def model_cofficient(input_var,output_var):
    linear_model.fit(X=input_var,y=output_var)
    return [linear_model.coef_,linear_model.intercept_]
```

```python
variable='bedrooms'
data_type=houseprice_train
input_var = data_type[variable].values.reshape(len(data_type[variable]),1)
output_var = data_type['price']
```

```python
model_cofficient(input_var,output_var)
```

```python
sqft=2650
price=(model_cofficient(input_var,output_var)[0][0]*sqft)+model_cofficient(input_var,output_var)[1]
print (price)
```
